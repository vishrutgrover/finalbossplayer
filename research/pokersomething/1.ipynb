```python
from xgboost import XGBRegressor
```

```python
from sklearn.metrics import r2_score, mean_squared_error
xgb_regressor = XGBRegressor()
# Fit the XGBoost regressor to the training data
xgb_regressor.fit(X_train, y_train)

# Predict on the training set
y_train_pred = xgb_regressor.predict(X_train)

# Calculate R-squared for training set
r2_train = r2_score(y_train, y_train_pred)

# Calculate Mean Squared Error (MSE) for training set
mse_train = mean_squared_error(y_train, y_train_pred)

# Predict on the test set
y_test_pred = xgb_regressor.predict(X_test)

# Calculate R-squared for test set
r2_test = r2_score(y_test, y_test_pred)

# Calculate Mean Squared Error (MSE) for test set
mse_test = mean_squared_error(y_test, y_test_pred)

# Print the results
print("Training R-squared:", r2_train)
print("Training MSE:", mse_train)
print("Test R-squared:", r2_test)
print("Test MSE:", mse_test)
```

```python
import math
y_pred = xgb_regressor.predict(X_test)
for i in range(len(y_pred)):
    if y_pred[i] < (2*math.ceil(y_pred[i])-1) / 2:
        y_pred[i] = math.floor(y_pred[i])
    else:
        y_pred[i] = math.ceil(y_pred[i])
print(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
y_test_pred = xgb_regressor.predict(X_test)

accuracy = accuracy_score(y_test, y_test_pred)

# Print the accuracy
print("Accuracy:", accuracy)
```
